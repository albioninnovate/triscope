```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

## Parameters 

```python
# Diameter of mirror in mm 

class telescope():
    
    def __init__(self,mirror_primary_diameter, 
                 mirror_primary_focal_lenght, 
                 eye_piece_diameter, 
                 eye_piece_focal_length):
        
        self.mirror_primary_diameter = mirror_primary_diameter
        self.mirror_primary_focal_lenght = mirror_primary_focal_lenght
       
        self.focal_ratio = mirror_primary_focal_lenght / mirror_primary_diameter
        
        self.eye_piece_diameter = eye_piece_diameter
        self.eye_piece_focal_length = eye_piece_focal_length
```

```python
t1 = telescope(115,600,9,50)
```

## Resolving Power 
in arcseconds 
 R = 120 / Diametre

```python
D =  t1.mirror_primary_diameter

def resolving_pwr_calc(D):
    """
    The smallest separation between two stars that can possibly be distinguished with the scope. 
The bigger the diameter of the objective, DO, the tinier the detail 
    """
    resolving_pwr = 120 / D
    
    return resolving_pwr 
```

```python
x = np.arange(100,D*2, 50)
y = resolving_pwr_calc(x)

plt.plot(x,y, label = "Resoluton Power")

x2 = [100,D*2]
y2 = [2.8,2.8]

plt.plot(x2,y2, label="Epsilon Lyrae ")

plt.title('Resolution Power')
plt.xlabel('Mirror Diameter (mm)')
plt.ylabel('Resolution (arc-sec)')
plt.legend()
```

```python
 # Star Magnitude Limit 
"""  Faintest Star Magnitude the Scope Can See """
    
def star_mag_limit_calc(D):
    
    star_mag_limit = 2 + 5 * math.log10(D)
    
    return star_mag_limit

star_mag_limit_calc(D)
```

```python
x = np.arange(100,D*2, 50)
y = np.vectorize(star_mag_limit_calc)

plt.plot(x,y(x), label = "Star Magnitude Limit")

#x2 = [100,D*2]
#y2 = [2.8,2.8]

#plt.plot(x2,y2, label="Epsilon Lyrae ")

plt.title('Star Magnitude Limit')
plt.xlabel('Mirror Diameter (mm)')
plt.ylabel('Star Magnitude Limit (magnatude)')
plt.legend()
```

```python
# Magnification

def magnification(objective_fl,eye_piece_fl): 
    """
    The apparent increase in size of an object when looking through the telescope, 
    compared with viewing it directly
    """
    magnification = objective_fl/eye_piece_fl
    
    return magnification
```
